Data Cleaning

dropping columns

```python
df.dropna(subset=["Rating"], inplace = True)
```

```python
df.dropna(subset=['Actor 1','Actor 2','Actor 3','Director','Genre'],inplace=True)
```

Convertion of datatypes

```python
#removing min from duration to keep it as int.
df['Duration'] = df['Duration'].str.strip('min')
```

# EDA(Exploratory Data Analysis)

```python
# Top 10 movies on basis of rating
top_movie = df.loc[df['Rating'].sort_values(ascending=False)[:10].index]
top_movie
```

Distribution of Top 10 movies wrt Year

```python
sns.histplot(data=top_movie, x="Year",hue="Rating", multiple="stack")
plt.title('Distribution of Top 10 Movies', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Rating', fontsize=14)
plt.tight_layout()
plt.show()
```

Genre movies

```python
genre = df['Genre'].value_counts().reset_index()
genre.columns = ['Genre', 'Count']
top_genres = genre.head(10)
top_genres
```

```python
plt.figure(figsize=(6, 6))
plt.pie(top_genres['Count'], labels=top_genres['Genre'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Movie Genre Distribution', fontsize=16)
plt.axis('equal')
plt.show()
```

Distribution of Top directors by average rating

```python
# Group the data by director and calculate the average rating
dir_avg = df.groupby('Director')['Rating'].mean().reset_index()
dir_avg = dir_avg.sort_values(by='Rating', ascending=False)
top_dir_names = dir_avg.head(10)
top_dir_names
#this gives directors based on their rating
```

```python
plt.figure(figsize=(13, 6))
sns.barplot(data=top_dir_names, x='Rating', y='Director', palette='viridis')
plt.title('Ranking directors based on their average ratings.', fontsize=20)
plt.xlabel('Average Rating', fontsize=17)
plt.ylabel('Director', fontsize=17)
plt.show()
```

top actors by number of movie

```python
actor_n = df['Actor 1'].value_counts().reset_index()
actor_n.columns = ['Actor', 'MovieCount']
top_actors = actor_n.head(10)
top_actors
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=top_actors, x='MovieCount', y='Actor', orient='h')
plt.title('Top Actors by Number of Movies', fontsize=16)
plt.xlabel('Number of Movies', fontsize=14)
plt.ylabel('Actor', fontsize=14)
plt.show()
```

 number of movie released every year

```python
yearly_counts = df['Year'].value_counts().reset_index()
yearly_counts.columns = ['Year', 'MovieCount']
yearly_movie_counts = yearly_counts.sort_values(by='Year')
yearly_movie_counts
```

```python
plt.figure(figsize=(13, 7))
sns.lineplot(data=yearly_movie_counts, x='Year', y='MovieCount')
plt.title('Number of Movies Released Every Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Movies Released', fontsize=14)
plt.show()
```

 Number of movies each genre

```python
df['Genre'] = df['Genre'].str.split(',')

# Create a new DataFrame with one row for each genre
genre_df = df.explode('Genre')
genre_df
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=genre_df, x='Genre', order=genre_df['Genre'].value_counts().index , palette='dark')
plt.title('No.of Movies for Each Genre', fontsize=16)
plt.xlabel('Number of Movies', fontsize=14)
plt.ylabel('Genre', fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

# Featured Engineering

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
trans_data = df.drop(['Name'],axis=1)
# Transforming Director columns
trans_data['Director'] = labelencoder.fit_transform(df['Director'])
# Transforming Actors Columns
trans_data['Actor 1'] = labelencoder.fit_transform(df['Actor 1'])
trans_data['Actor 2'] = labelencoder.fit_transform(df['Actor 2'])
trans_data['Actor 3'] = labelencoder.fit_transform(df['Actor 3'])
trans_data['Genre'] = labelencoder.fit_transform(df['Genre'].apply(lambda x: ', '.join(x)))
trans_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
sc_data = scaler.fit_transform(trans_data)
sc_df = pd.DataFrame(sc_data,columns=trans_data.columns)
sc_df.head()
```

Examine Correlation of Data

```python
# correlation
corr_df = trans_data.corr(numeric_only=True)
corr_df['Rating'].sort_values(ascending=False)
```

```python
sns.heatmap(corr_df,annot=False,cmap="coolwarm")
```

# Model Building

```python
# Import modeling Libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
# keep the data except Rating data
X = trans_data.drop(['Rating'],axis=1)
# keep only Rating data
y = trans_data['Rating']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
X_test = np.array(X_test)
y_pred = model.predict(X_test)
y_pred
```

```python
print(y_test)
```

```python
print('R2 score: ',r2_score(y_test,y_pred))
print('Mean squared error: ',mean_squared_error(y_test,y_pred))
print('Mean absolute error: ',mean_absolute_error(y_test,y_pred))
```
